### Importing

```python
import pandas as pd
import numpy as np
```

```python
adult_income = pd.read_csv("adult-income.csv")
```

```python
for col in adult_income.columns:
    print(col)
```

```python
adult_income = adult_income.rename(columns={"39": "Age", " State-gov": "Workclass", " 77516": "Final_weight", 
                             " Bachelors": "Education", " 13": "Education_num", " Never-married": "Marital_status", 
                             " Adm-clerical": "Occupation", " Not-in-family": "Relationship", " White": "Race", 
                             " Male": "Gender", " 2174": "Capital_gain", " 0": "Capital_loss", 
                             " 40": "hours-per-week", " United-States": "Native_country", 
                             " <=50K": "Income_bracket"})
```

### Sorting a DataFrame - sort_values()

```python
df_sorted = adult_income.sort_values(by="Education_num", axis=0, ascending=False)
df_sorted.head()
```

```python
df_sorted.tail()
```

```python
df_sorted = adult_income.sort_values(by=["Education_num","hours-per-week"], axis=0, ascending=False)
df_sorted.head(12)
```

```python
df_sorted = adult_income.sort_values(by=["Education_num","hours-per-week"], axis=0, ascending=[False, True])
df_sorted.head(12)
```

### Top Values - nlargest() and nsmallest()

```python
df_largest = adult_income.nlargest(10, "Capital_gain")
df_largest
```

```python
df_smallest = adult_income.nsmallest(5, "hours-per-week")
df_smallest
```
